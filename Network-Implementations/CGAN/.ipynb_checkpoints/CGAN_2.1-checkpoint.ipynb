# CGAN 1.0

This is an implementation of Conditional GAN based on dataset MNIST. 
Note that I replace most of the fully connected layers with transposed convolution layers to have a better performance.

```python
import argparse
import os
import numpy as np
import math

import torchvision.transforms as transforms
from torchvision.utils import save_image

from torch.utils.data import DataLoader
from torchvision import datasets
from torch.autograd import Variable

import torch.nn as nn
import torch.nn.functional as F
import torch

os.makedirs("images_1.0", exist_ok=True)
```

```python
class OPT():
    def __init__(self):
        self.n_classes = 10
        self.n_epochs = 200
        self.batch_size = 64
        self.lr = 0.0002
        self.b1 = 0.5
        self.b2 = 0.999
        self.n_gpu = 2
        self.latent_dim = 100
        self.img_size = 32
        self.channels = 1
        self.sample_interval = 400

opt = OPT()

print(opt.n_epochs)
```

```python
img_shape = (opt.channels, opt.img_size, opt.img_size)

cuda = True if torch.cuda.is_available() else False

print(img_shape)
print(cuda)
```

```python
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        
        self.label_emb = nn.Embedding(opt.n_classes, opt.n_classes)
        
        def block(in_feature, out_feature, normalize=True):
            layers = [nn.Linear(in_feature, out_feature)]
            if normalize:
                layers.append(nn.BatchNorm1d(out_feature))
            layers.append(nn.LeakyReLU(0.2, inplace=True))
            return layers
        
        self.model = nn.Sequential(
            *block(opt.latent_dim + opt.n_classes, 128, normalize=False),
            *block(128, 256),
            *block(256, 512),
            *block(512, 1024),
            nn.Linear(1024, int(np.prod(img_shape))),
            nn.Tanh()
        )
        
    def forward(self, noise, labels):
        '''
        # noise: [batch_size, latent_dim]
        # labels: [batch_size, ] each number is in range of [0, 9] indicating the class 
            # self.label_emb(labels): [batch_size, n_classes]
        '''
        x = torch.cat((noise, self.label_emb(labels)), 1)
        img = self.model(x) # [batch_size, img_size * img_size * channels]
        img = img.view(img.size(0), *img_shape)
        return img
```

```python
test_noise = torch.randn(64, 100)
test_labels = torch.LongTensor(np.zeros(64))

print(test_noise.size())
print(test_labels.size())

generator = Generator()

test_out = generator(test_noise, test_labels)
print(test_out.size())
```

```python
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        
        self.label_emb = nn.Embedding(opt.n_classes, opt.n_classes)
        
        self.model = nn.Sequential(
            nn.Linear(opt.n_classes + int(np.prod(img_shape)), 512),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(512, 256),
            nn.Dropout(),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(256, 128),
            nn.Dropout(),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(128, 1),
            nn.Sigmoid()
        )
    
    def forward(self, img, labels):
        '''
        # img: [batch_size, channels, img_size, img_size]
        # labels: [batch_size,]
        '''
        x = img.view(img.size(0), -1)
        x = torch.cat((x, self.label_emb(labels)), 1)
        out = self.model(x)
        return out
```

```python
test_img = torch.randn(64, *img_shape)
test_labels = torch.LongTensor(np.zeros(64))

print(test_img.size())
print(test_labels.size())

discriminator = Discriminator()

test_out = discriminator(test_img, test_labels)
print(test_out.size())
```

```python
dataloader = torch.utils.data.DataLoader(
    datasets.MNIST(
        "./dataset",
        train=True,
        download=True,
        transform=transforms.Compose(
            [transforms.Resize(opt.img_size), transforms.ToTensor(), transforms.Normalize([0.5], [0.5])]
        ),
    ),
    batch_size=opt.batch_size,
    shuffle=True,
)
print(dataloader)
```

```python
print(torch.cuda.current_device())
print(torch.cuda.device_count())
```

```python
adversarial_loss = torch.nn.MSELoss()

device = torch.device("cuda:3" if cuda else "cpu")
generator.to(device)
discriminator.to(device)
adversarial_loss.to(device)

FloatTensor = torch.cuda.FloatTensor if cuda else torch.FloatTensor
LongTensor = torch.cuda.LongTensor if cuda else torch.LongTensor

optimizer_G = torch.optim.Adam(generator.parameters(), lr=opt.lr, betas=(opt.b1, opt.b2))
optimizer_D = torch.optim.Adam(discriminator.parameters(), lr=opt.lr, betas=(opt.b1, opt.b2))
```

```python
fixed_noise = Variable(FloatTensor(np.random.normal(0, 1, (10 ** 2, opt.latent_dim)))).to(device)
fixed_label =  Variable(LongTensor(np.array([num for _ in range(10) for num in range(10)]))).to(device)

def save_images(batches_done):
    gen_imgs = generator(fixed_noise, fixed_label)
    save_image(gen_imgs.data, "images_1.0/%d.png" % batches_done, nrow=10, normalize=True)
```

```python
for epoch in range(opt.n_epochs):
    for i, (imgs, labels) in enumerate(dataloader):
        
        batch_size = imgs.shape[0]
        
        # Set Up
        valid = Variable(FloatTensor(batch_size, 1).fill_(1.0), requires_grad=False).to(device)
        fake = Variable(FloatTensor(batch_size, 1).fill_(0.0), requires_grad=False).to(device)
        
        real_imgs = Variable(imgs.type(FloatTensor).to(device))
        labels = Variable(labels.type(LongTensor).to(device))
        
        #-----------------
        # Train Generator
        #-----------------
        
        optimizer_G.zero_grad()
        
        z = Variable(FloatTensor(np.random.normal(0, 1, (batch_size, opt.latent_dim)))).to(device)
        gen_labels = Variable(LongTensor(np.random.randint(0, opt.n_classes, batch_size))).to(device)
        
        gen_imgs = generator(z, gen_labels)
        
        validity = discriminator(gen_imgs, gen_labels)
        g_loss = adversarial_loss(validity, valid)
        
        g_loss.backward()
        optimizer_G.step()
        
        #---------------------
        # Train Discriminator
        #---------------------
        
        optimizer_D.zero_grad()
        
        validity_real = discriminator(real_imgs, labels)
        d_real_loss = adversarial_loss(validity_real, valid)
        
        validity_fake = discriminator(gen_imgs.detach(), gen_labels)
        d_fake_loss = adversarial_loss(validity_fake, fake)
        
        d_loss = d_real_loss + d_fake_loss
        
        d_loss.backward()
        optimizer_D.step()
        
        print(
            "[Epoch %d/%d] [Batch %d/%d] [D loss: %f] [G loss: %f]"
            % (epoch + 1, opt.n_epochs, i, len(dataloader), d_loss.item(), g_loss.item())
        )
        
        batches_done = epoch * len(dataloader) + i
        if batches_done % opt.sample_interval == 0:
            save_images(batches_done)
```
